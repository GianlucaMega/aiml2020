# Starting images conversion
Both VOC2007 and VOC2012 trainval splits are processed by the trained CycleGAN in order to obtain domain transferred images
<br>DT is made with no preprocessing of the images in order to maintain the original dimensions to be matched by the annotations

```
%cd /content/pytorch-CycleGAN-and-pix2pix
!python test.py --dataroot datasets/vocconv/voc2007 --name vocconv --model test --no_dropout --num_test 5012 --preprocess none
```

Since the CycleGAN produces both the real and the fake images in png format a batch renaming and conversion of the transferred images is needed. Images are also moved in the original VOCdevkit folder substituting the original ones.
<br>This is made for both VOC2007 and VOC2012 

```
%cd /content
!rm -r VOCdevkit/VOC2007/JPEGImages/*
!mv pytorch-CycleGAN-and-pix2pix/results/vocconv/test_latest/images/* VOCdevkit/VOC2007/JPEGImages/
%cd VOCdevkit/VOC2007/JPEGImages
!rename 's/_fake//g' *
!rm *real*
!mogrify -format jpg *.png
!rm *.png
```

```
%cd /content/pytorch-CycleGAN-and-pix2pix
!python test.py --dataroot datasets/vocconv/voc2012 --name vocconv --model test --no_dropout --num_test 11541 --preprocess none
```
